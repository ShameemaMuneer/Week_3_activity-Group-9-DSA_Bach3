## Answer

### **1. Displaying the Glucose level and the Age for the samples having the index ranging from 45 to 60**

```python
#Reading the data set to python environment
data_diab=pd.read_csv('diabetes_model.csv')
data_diab
```

```python
#slicing and displaying Glucose and Age column having the index range:45-60 ,from the data set
data_slice=data_diab[['Glucose','Age']][45:61]
data_slice
```

### **2. Counting and displaying patients having the blood pressure level above 120 in the data set**

```python
bp_above_120=data_diab[data_diab['BloodPressure']>120]
bp_above_120
```

```python
#only one patient is there having blood pressure level above 120
```

### **3. Finding and displaying the minimum and maximum value of insulin level in the data**

**one method**

```python
#finding min value of insulin
min_insulin=data_diab['Insulin'].min()
min_insulin
```

```python
#finding max value of insulin
max_insulin=data_diab['Insulin'].max()
max_insulin
```

**Another method using describe()**

```python
data_diab.describe()
```

### **4.Plotting Histogram of 'Age'**  

```python
#creating a Histogram by assigning its colour,density,size,label,etc
plt.figure(figsize=(8,6))
plt.hist(data_diab['Age'],color='green',rwidth=0.9,density='true')
plt.title('Histogram: Distribution of Age',fontsize=18)
plt.xlabel('Age',fontsize=12)
plt.ylabel('count',fontsize=12)
plt.grid()
```

### **5. Plotting scatter plot of "BMI" vs "Glucose"**

```python
#creating a Scatter plot by assigning its colour,marker,size,label,etc
plt.figure(figsize=(8,6))
plt.scatter(data_diab['BMI'],data_diab['Glucose'],s=12,c='orange',marker='*')
plt.title('Scatter Plot: BMI vs Glucose',fontsize=18)
plt.xlabel('BMI',fontsize=12)
plt.ylabel('Glucose',fontsize=12)
```
